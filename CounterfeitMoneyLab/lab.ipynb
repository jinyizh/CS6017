Use the ``sklearn`` ``feature_selection`` ``SelectPercentile`` function using the f_classif statistic to quantify which features are likely to be best for use with classification

```python
predictor_features = dataframe.columns[:-1]
selector = SelectPercentile(f_classif)

X_new = selector.fit_transform(dataframe[predictor_features].values, dataframe.is_real.values)
```
