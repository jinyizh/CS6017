```python
import numpy as np
import pandas as pd
import matplotlib.pyplot as plt
import statsmodels.formula.api as sm
```

## Dumb KNN

```python
# plot parts of data to make sense of it
dumbknn = pd.read_csv("spatialDatastructuresStarterCode/dumb_knn.csv")
pd.plotting.scatter_matrix(dumbknn)
figure = plt.gcf()
figure.set_size_inches(10, 10, forward=True)
plt.show()
```

```python
# perform regression
dumbReg = sm.ols(formula="time ~ n + k + dimension", data=dumbknn).fit()
dumbReg.summary()
```
